```python
import numpy as np
import pandas as pd
#from collections import Counter
import networkx as nx
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
#G = nx.read_weighted_edgelist("parte2")
G = nx.barabasi_albert_graph(10, 3)

nx.draw(G, with_labels=True)
```

```python
# degree of each node
# link number that each node has
degrees = [deg for node, deg in nx.degree(G)]
```

```python
print(degrees)
```

```python
# kmin - minimum degree
kmin = np.min(degrees)

print(kmin)
```

```python
# kmax - maximum degree
kmax = np.max(degrees)

print (kmax)
```

```python
# kavg - average degree
kavg = np.mean(degrees)

print(kavg)
```

```python
# Degree distribution: helps us understand connectivity trends in networks and 
# how edges are distributed among nodes (does everyone have similar number of connections, 
# or do we have hubs, nodes with significantly higher number of connections?)


def degree_distr(net):
    degrees = dict(net.degree()) 
    hist = list(Counter(degrees.values()).items()) 
    hist.sort(key=lambda x:x[0])
    hist = np.array(hist)
    return hist
```
