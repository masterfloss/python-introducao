```
#Uma lista armazena uma série de items numa ordem específica
#Pode-se aceder a cada item utilizando um indice ou através de ciclo
#As listas são mutáveis

# Construir uma lista
lista_compras = ['batatas', 'cenouras','bacalhau', 'couves']
#Ou então
lista_compras = list(('batatas', 'cenouras','bacalhau', 'couves'))
```

```
#Operações com listas

# Obter o primeiro elemento da lista 
lista_compras[0]


```

```
# Obter o último elemento da lista 
lista_compras[-1]
```

```
# Iterar através de uma lista
for compra in lista_compras:
    print(compra)
```

```
# Adicionar items a lista 
filmes =[]
filmes.append('Vice')
filmes.append('Green Book')
filmes.append('Roma')
filmes.append('A Star Is Born')
print (filmes)
```

```
#Remover Elementos
filmes.pop()
filmes.remove('A Star Is Born')
del films[1]
print (filmes)
```

```
#Apagar lista
clear(filmes)
```

```
#Compreensão de lista

# List comprehensions 
squares = [x**2 for x in range(1, 11)]
squares
```
