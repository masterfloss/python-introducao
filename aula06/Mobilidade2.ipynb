```python
import numpy as np
import pandas as pd
from collections import Counter
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# criar uma rede sem elementos
G = nx.Graph()
```

```python
for a,b, c in zip(s.SendingCountry, s.ReceivingCountry, s.DurationInMonths): 
    G.add_edge(a,b, weight= c) 
    
```

```python
#nx.draw(G)

#nx.draw(G, with_labels=True)

nx.draw_circular(G)
```

```python
# grau de ns, grau mdio da rede, distribuio da rede
degrees = [deg for node, deg in nx.degree(G)]
```

```python
# kmin - grau minimo
kmin = np.min(degrees)

print(kmin)
```

```python
# kmax - grau maximo
kmax = np.max(degrees)

print (kmax)
```

```python
# kavg - grau mdio da rede
kavg = np.mean(degrees)

print(kavg)
```

```python
#Distribuio em graus: ajuda-nos a entender as tendncias de conectividade em redes e
#como os arcos so distribudas entre ns (todos tem um nmero similar de conexes,
#ou temos hubs, ns com um nmero significativamente maior de conexes?)

def degree_distr(net):
    degrees = dict(net.degree()) 
    hist = list(Counter(degrees.values()).items()) 
    hist.sort(key=lambda x:x[0])
    hist = np.array(hist)
    return hist
```

```python
degree_distr(G)
```

```python
# triangulos
nx.triangles(G)
```

```python
#  coefieicente de clustering de um n 
nx.clustering(G)
```

```python
# coeficiente de clustering the todos os ns (retorna um dicionrio)

nx.clustering(G)
```

```python
# coediciente de clustering da rede
cc = nx.clustering(G)
avg_clust = sum(cc.values()) / len(cc)
print(avg_clust)
```

```python
# centralidade de betweenness da rede
nx.betweenness_centrality(G)
```

```python
# proximidade centralidade da rede 
nx.closeness_centrality(G)
```

```python
# eigenvector centralidade da rede
nx.eigenvector_centrality(G)
```

```python
# grau de centralidade
nx.degree_centrality(G)
```

```python
# nmero de componentes ligados
nx.number_connected_components(G)
```

```python
# obter o n ligado a um componente  
nx.node_connected_component(G, "PT")
```

```python
# Assortatividade (Assortativity)
# Coeficiente de correlao de Pearson [-1; 1]
# Redes sociais so altamente combinadas (homofilia): ns de alto grau ligados a outros ns de alto
# grau tecnolgicos so disassortative: ns de alto grau se conectam a ns de baixo grau

nx.degree_assortativity_coefficient(G)
```
