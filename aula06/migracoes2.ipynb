```python
import numpy as np
import pandas as pd
from collections import Counter
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# criar uma rede sem elementos
#G = nx.Graph()
G = nx.DiGraph()
```

```python
for a,b, c in zip(df.origem, df.destino, df.numeros):
    if(c>1000000):
        G.add_edge(a,b, weight= c) 
    
```

```python
import matplotlib.pyplot as plt
pos = nx.spring_layout(G,dim=2,scale=1.0)
# nodes
nx.draw_networkx_nodes(G, pos, node_size=200)

elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.5]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 0.5]

# edges
nx.draw_networkx_edges(G, pos, edgelist=elarge,
                       width=1)
nx.draw_networkx_edges(G, pos, edgelist=esmall,
                       width=1, alpha=0.5, edge_color='b', style='dashed')

# labels
nx.draw_networkx_labels(G, pos, font_size=8, font_family='sans-serif', font_color='g',)

plt.axis('on')
plt.show()
```

```python
# grau de nós, grau médio da rede, distribuição da rede
degrees = [deg for node, deg in nx.degree(G)]
```

```python
# kmin - grau minimo
kmin = np.min(degrees)

print(kmin)
```

```python
# kmax - grau maximo
kmax = np.max(degrees)

print (kmax)
```

```python
# kavg - grau médio da rede
kavg = np.mean(degrees)

print(kavg)
```

```python
#Distribuição em graus: ajuda-nos a entender as tendências de conectividade em redes e
#como os arcos são distribuídas entre nós (todos tem um número similar de conexões,
#ou temos hubs, nós com um número significativamente maior de conexões?)

def degree_distr(net):
    degrees = dict(net.degree()) 
    hist = list(Counter(degrees.values()).items()) 
    hist.sort(key=lambda x:x[0])
    hist = np.array(hist)
    return hist
```

```python
degree_distr(G)
```

```python
# Assortatividade (Assortativity)
# Coeficiente de correlação de Pearson [-1; 1]
# Redes sociais são altamente combinadas (homofilia): nós de alto grau ligados a outros nós de alto
# grau tecnológicos são disassortative: nós de alto grau se conectam a nós de baixo grau

nx.degree_assortativity_coefficient(G)
```
