```
#Sets ( ou conjuntos) 
#São estruturas disponíveis no Python, utilizadas para representar coleções não ordenadas de elementos únicos. 

s = {1, 2, 3, 4}
print (s)
```

```
# Nos Conjuntos podem-se realizar um conjunto de operações típicas dos conjuntos da matemática, tais como:
# - Interseção
# - União
# - Diferença
# - Diferença simétrica
```

```
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

```

```
print (a.union(b))
```

```
print (a.intersection(b))
```

```
print (b.difference(a))
```

```
print (a.symmetric_difference(b))
```
