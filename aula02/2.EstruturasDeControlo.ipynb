```
# Um simples if
idade=23
if idade >= 18: 
	print("Podes votar!") 

```

```
#Declaração If-elif-else
if idade < 3: 
	preco_bilhete= 0 
elif idade < 18: 
	preco_bilhete = 5 
else: preco_bilhete = 10
```

```
# Repetição
# Um simples loop while
valor_atual= 1 
while valor_atual <= 5: 
	print(valor_atual) 
	valor_atual += 1
```

```
# O ciclo termina como resultado do 
# valor introduzido pelo utilizador

msg = ''
while msg != 'sair': 
	msg = input(“Qual a tua mensagem? ") 
	print(msg)
```
