```
#Os dicionários armazenam ligações entre peças de informação
#Cada item de um dicionário é um par chave-valor
#Os valores não são repetiveis

fruta = {1: 'laranja', 2: 'maça', 3:'pera',4:'uva',5:'pessego'}


```

```
#adicionar um novo par chave-valor 

fruta[10] = 'romã'
```

```
#Iterar através de par chave-valor

for chave , valor in fruta.items():
    print('a fruta ' + str(chave) +' é ' + valor)
```

```
#Iterar através da chave
for chave in fruta.keys(): 
   print(str(chave) + ' é fruta')
```

```
# Iterar através dos valores
for valor in fruta.values(): 
   print(valor + ' é fruta')
```
